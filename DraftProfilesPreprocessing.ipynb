```python
import pandas as pd
from google.colab import auth
from google.cloud import storage
import io
import re
auth.authenticate_user()
```

```python
client = storage.Client(project='cmse-381')
bucket = client.get_bucket('stats_draftprospects')
blob = bucket.blob('nfl_draft_prospects.csv')
content = blob.download_as_string()
prospects = pd.read_csv(io.BytesIO(content))
blob = bucket.blob('nfl_draft_profiles.csv')
content = blob.download_as_string()
profiles = pd.read_csv(io.BytesIO(content))
```

```python
prospects
```

```python
profiles
```

```python
# drop columns from profiles that are columns in prospects besides player_name
drop = []
for column in profiles.columns:
    if column in prospects.columns and column != 'player_name':
        drop.append(column)
profiles.drop(columns=drop, inplace=True)

```

```python
profiles
```

```python
# drop all rows where all text columns are NaN
profiles.dropna(subset=['text1', 'text2', 'text3', 'text4'], how='all', inplace=True)
```

```python
# drop text3 and text4 and drop all rows where both text1 and text 2 are NaN
profiles.drop(columns=['text3', 'text4'], inplace=True)
profiles.dropna(subset=['text1', 'text2'], how='all', inplace=True)

```

```python
# if text1 is NaN, replace with text2 in new column 'draft_profile'

profiles['draft_profile'] = profiles['text1'].fillna(profiles['text2'])
```

```python
profiles
```

```python
# remove phrase <b>What he brings:</b>
profiles['draft_profile'] = profiles['draft_profile'].apply(lambda x: re.sub(r'<b>What he brings:</b>', '', x))
```

```python
# merge profiles and prospects into new df, draft_profiles
draft_profiles = pd.merge(profiles, prospects, on='player_name')
```

```python
draft_profiles
```

```python
draft_profiles.columns
```

```python
draft_profiles.drop(columns=['alt_player_id', 'school_logo', 'text1', 'text2', 'player_id', 'position', 'school_name', 'school',
                             'link', 'traded', 'trade_note', 'team', 'team_logo_espn', 'guid', 'pos_rk', 'ovr_rk', 'grade'], inplace = True)
```

```python
draft_profiles
```

```python
# drop when pick equals NaN
draft_profiles.dropna(subset=['pick'], inplace=True)
```

```python
draft_profiles
```

```python
# remove duplicate names
draft_profiles.drop_duplicates(subset=['player_name'], inplace=True)
```

```python
# reindex
draft_profiles.reset_index(drop=True, inplace=True)
```

```python
draft_profiles
```

```python
# export to gcs storage bucket
draft_profiles.to_csv('draft_profiles.csv', index=False)
blob = bucket.blob('draft_profiles.csv')
blob.upload_from_filename('draft_profiles.csv')
```
